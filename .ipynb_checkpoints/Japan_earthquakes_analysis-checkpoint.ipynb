extreme volcanic eruption for which data is available can be categorized as Catastrophe Type II according to the

are studied. A multidimensional scale for volcanic eruptions is investigated. Intensity, fatalities, affected
population, impacted region, cost of damage, and GDP per capita, are some factors that can be considered to
determine the severity level. Relationships among the factors are also considered. An analysis is conducted to
identify the specific factors to be considered in the multidimensional scale. The extreme values of known

### Introducion 
Volcanoes can be classified
according to the level of activity as follows: active (presently erupting), dormant (not presently erupting but could at future date), and extinct (no eruptions in

Special types of volcanoes are also distinguished, such as super volcanoes or hot spots.
Historical evidence of super volcanoes does not currently exist; however, evidence for these massive
phenomena have been observed in the geological record.<br>
 Examples of some hotspots are Hawaii, Yellowstone National Park (United States), Iceland, Samoa, and Bermuda.

### Handling the dates

Because the standard date library in Python cannot handle UTC dates before 1970 , we are using two additional libraries as follows:
- convertdate - for handling dates before introduction of Julian calendar <br>
- skyfield -  for handling gregorian calendar dates before year 1582<br> 

    Because date 1 of january was set additionaly  , and after the convertion there was 2 days offset , in case of 1 January we set the day to be 3 th to avoid wrong conversion of the year. Usually it was 1 year behind.

```python
def convert_gregorian_date(year,month,day):
    """Handles the date convertion and formatting of the dates"""
    if year < 1582:
    # The date 01/01 is artifitially set. We set the day to 03 to prevent the convertion to go 1 year behind
        if (month == 1) & (day == 1):
            day = 3
        date = julian.to_gregorian(year,month,day)
        return str(date[2]) +'/'+str(date[1]) +'/'+ str(date[0])
    else:
       return ts.utc(year,month,day).utc_strftime('%d/%m/%Y')
```

```python
def get_common_eruptions(name):
    """If there isn't any significant eruptions we get the events data from Kaggle only"""
    return merged_events[merged_events['volcano name'] == name].fillna(value = {'vei':0})
```

```python
def get_significant_eruptions(name):

    """Separate the most significant events for the current volcano """
    v_significant_eruptions = significant_eruptions[significant_eruptions['NAME'] == name]
    v_significant_eruptions = v_significant_eruptions[['YEAR', 'VEI', 'DEATHS_TOTAL', 'COMMENTS']]
    v_significant_eruptions = v_significant_eruptions.sort_values(by=['YEAR'], ascending=False).reset_index()
    v_significant_eruptions = v_significant_eruptions.drop(columns=['index'])
    v_significant_eruptions = v_significant_eruptions.rename(columns={'YEAR':'year','VEI':'vei',
                                                                      'DEATHS_TOTAL':'fatalities count',
                                                                      'COMMENTS':'comments'})
    v_significant_eruptions = v_significant_eruptions.dropna(subset=['vei','comments'], how='all')
    v_significant_eruptions = v_significant_eruptions.dropna(subset=['fatalities count'], how='any')
    v_significant_eruptions['fatalities count'] = v_significant_eruptions['fatalities count'].astype('int64')
    return v_significant_eruptions
```

```python
def get_event_wo_fatalities(event_set):
    """Clearing and returning the non fatality records"""

    event_set = event_set.drop(columns=['start_year','eruption category','volcano number'])
    return event_set.drop_duplicates(['eruption number']).reset_index()
```

```python

def construct_map(names,ll_lon, ll_lat, ur_lon, ur_lat):
    """To be used for displaying the certain location
        accepts:
        - array of the names of the volcanoes which should be displayed
        - lower left corner , and upper right corner by their geographical longitude and latitude"""

    plt.figure(figsize=(12,6))

    """Creating basemap using coordinate system WGS84"""
    map = Basemap(ll_lon, ll_lat, ur_lon, ur_lat, epsg=3857)

    """Using arcgisimage method to show more detailed image with higher resolution"""
    map.arcgisimage(service='ESRI_Imagery_World_2D', xpixels = 1500, verbose= True,zorder=2)

    volcanoes = []
    for n in names:
        volcanoes.append(volcano_data[volcano_data['name'] == n])

    """Prepearing the coordinated for the markers of the two volcanoes"""
    lons = []
    lats = []
    for v in volcanoes:
        lons.append(v['longitude'].values)
        lats.append(v['latitude'].values)
    lonlats = [lons,lats]

    """Mark the volcano location """
    map.scatter(lonlats[0], lonlats[1],color="red", marker="o",latlon=True)

    """Transform label coordinates and display the label itself"""
    for i in range(len(names)):
        cx, cy = map(lonlats[0][i], lonlats[1][i])
        plt.text(cx, cy, names[i], fontsize=12,fontweight='bold',ha='left',va='bottom',color='w')

    plt.show()
```

```python
"""Grouping the data by volcano name to fill the mean value of Explosivity Index (VEI)"""
eruption_by_volcano = merged_events.groupby(['volcano name'])['vei'].mean().dropna().astype(int)\
.rename('processed vei').reset_index().merge(merged_events)
```

```python
def get_fatatliy_eruptions(names):
    """Returning a combined dadtaframe with information about certain event and the casualties related to it """
    er = pd.DataFrame()

    for name in names:
        """Separete volcano's fatality eruptions"""
        v_eruptions = eruption_by_volcano[eruption_by_volcano['volcano name']==name]
        v_eruptions = v_eruptions.drop(columns=['volcano number'])
        v_fatality_eruptions = v_eruptions[v_eruptions['event_type'] == 'Fatalities'].reset_index()
        v_fatality_eruptions = v_fatality_eruptions.drop(columns=['index'])

        """Merging volcano's fatality eruption dataframe with another dataset which contains data about
        count of fatalities and devastations and some detailed descriptions about these events """
        v_fatality_eruptions = pd.merge(v_fatality_eruptions, get_significant_eruptions(name),
                                        left_on='start_year',right_on='year' ).drop_duplicates()
        v_fatality_eruptions = pd.merge(v_fatality_eruptions, volcano_data[['name','population within 5 km',
                                                                            'population within 10 km',
                                                                            'population within 30 km',
                                                                            'population within 100 km']],
                                        left_on='volcano name',right_on='name' ).drop_duplicates()
        v_fatality_eruptions = v_fatality_eruptions.fillna(value = {'vei_y':0})
        v_fatality_eruptions['vei_y'] = v_fatality_eruptions['vei_y'].astype('int64')
        v_fatality_eruptions = v_fatality_eruptions.rename(columns={'vei_y':'volcano explosivity index(vei)'})
        v_fatality_eruptions = v_fatality_eruptions.drop(columns=['vei_x','start_year','event date',
                                                                  'eruption number','eruption category'])
        er = er.append(v_fatality_eruptions,ignore_index=True)

    return er
```

## Volcanic Explosivity Index
The Volcanic Explosivity Index (VEI) is a relative measure of the explosiveness of volcanic eruptions. It was devised by Chris Newhall of the United States Geological Survey and Stephen Self at the University of Hawaii in 1982.

Volume of products, eruption cloud height, and qualitative observations (using terms ranging from "gentle" to "mega-colossal") are used to determine the explosivity value. The scale is open-ended with the largest volcanoes in history given magnitude 8. A value of 0 is given for non-explosive eruptions, defined as less than 10,000 m3 (350,000 cu ft) of tephra ejected; and 8 representing a mega-colossal explosive eruption that can eject 1.0×1012 m3 (240 cubic miles) of tephra and have a cloud column height of over 20 km (66,000 ft). The scale is logarithmic, with each interval on the scale representing a tenfold increase in observed ejecta criteria, with the exception of between VEI-0, VEI-1 and VEI-2.<br>

```python
volcano_names = fatalities_by_volcano['name'].values
all_fatatlity_records = get_fatatliy_eruptions(volcano_names)
all_fatatlity_records = all_fatatlity_records[(all_fatatlity_records['volcano explosivity index(vei)'] > 0) &
                                              (all_fatatlity_records['fatalities count'] > 0) ]
```

```python
sns.lmplot(x='volcano explosivity index(vei)', y='fatalities count', ci=None, data = all_fatatlity_records)
ax = plt.gca()
ax.set_title('Corelation between fatalities count and VEI')
```

```python
correlation = all_fatatlity_records['volcano explosivity index(vei)'].corr(all_fatatlity_records['fatalities count'])
print('Correlation between VEI and count of fatalities is {:.2f}'.format(correlation))
```

So here we have a positive moderate correlation between these variables. The more explosive is an eruption, the more casualties it can cause.

### Correlation between population in range near a volcano and the count of casualties

```python
correlation_5km = all_fatatlity_records['fatalities count'].corr(all_fatatlity_records['population within 5 km'])
correlation_10km = all_fatatlity_records['fatalities count'].corr(all_fatatlity_records['population within 10 km'])
correlation_30km = all_fatatlity_records['fatalities count'].corr(all_fatatlity_records['population within 30 km'])
correlation_100km = all_fatatlity_records['fatalities count'].corr(all_fatatlity_records['population within 100 km'])
d = {'population range':['population within 5 km','population within 10 km','population within 30 km',
                         'population within 100 km'],
        'correlation':[correlation_5km,correlation_10km,correlation_30km,correlation_100km]}
corr_frame = pd.DataFrame(data=d).round({'correlation':2})
corr_frame
```

So , we receive very weak correlations and we can conclude that according the data , the population around the volcanoes almost doesn't affect the count of casualties in case of eruption.
